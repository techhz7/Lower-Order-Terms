```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Demonstrating the impact of lower order terms
n_values = np.arange(1,40)
n_squared = n_values ** 2
n_plus_n_squared = n_values ** 2 + n_values
```

```python
# plotting the results
plt.plot(n_values,n_squared,label = 'n^2')
plt.plot(n_values,n_plus_n_squared,label= 'n^2 + n', linestyle= '--')
plt.xlabel('Input_Size(n)')
plt.ylabel('value')
plt.title('Impact of Lower-Order terms')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Demonstrating Lower Order Terms
n_values = np.arange(1,40)
n_cube = n_values ** 3
n_plus_cubed = n_cube + n_values
```

```python
# Plotting the results
plt.plot(n_values,n_cube, label='n^3')
plt.plot(n_values , n_plus_cubed,label='n^3 + n', linestyle ='--' )
plt.xlabel('Input Size(n)')
plt.ylabel('Value')
plt.title('Impact of Lower Order terms')
plt.legend()
plt.grid(True)
plt.show()
```
