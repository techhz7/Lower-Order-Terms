```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_values = np.arange(1,40)
n_squared = n_values ** 2
n_plus_n_squared = n_squared + n_values
```

```python
plt.plot(n_values,n_squared, label='n^2')
plt.plot(n_values,n_plus_n_squared, label='n^2 + n', linestyle='--')
plt.xlabel("Input Size(n)")
plt.ylabel("value")
plt.title("Impact Of Lower-Order terms")
plt.legend()
plt.grid(True)
plt.show()
```
