```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Demonstrating Cubic roots
s_values = np.arange(1,7000)
s_cubic = s_values ** (1/3)
s_minus_cubic_root = s_cubic - s_values
```

```python
plt.plot(s_values, s_cubic,label='s^(1/3)')
plt.plot(s_cubic,s_minus_cubic_root,label='s^(1/3) - S',linestyle='--')
plt.xlabel('Input Size s^(1/3)')
plt.ylabel('s-Value')
plt.title('Impact of Cubic Root')
plt.legend()
plt.grid(True)
plt.show()
```
